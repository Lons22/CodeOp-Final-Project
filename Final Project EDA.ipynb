```python
#Overall Primary
overall_primary = new_data[['parentesco1', 'male', 'female', 'instlevel3']].groupby(by=['parentesco1', 'instlevel3']).sum().reset_index()

tempdf = overall_primary[overall_primary['parentesco1']== 1] #access column and filtering on its value
 
newdf = pd.melt(tempdf, id_vars=['parentesco1', 'instlevel3'])
newdf=newdf.drop('parentesco1', axis = 1)

newdf

sns.barplot(x="variable", y="value", hue="instlevel3", data=newdf);

# 1=completed primary school, this graph shows that overall people have not completed primary school, of those that did, more males did compared to females.

```

```python
#What Percentage of HoH Completed Primary school by location? #instlevel3 = 1 means completed primary school

#1st Step: group and filter by HoH
Primary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel3']].groupby(by=['parentesco1', 'area1', 'instlevel3']).sum().reset_index()

tempdf = (Primary[Primary['parentesco1']== 1])
tempdf
#Now filter to only those who completed primary school

completed_only = (tempdf[tempdf['instlevel3']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel3', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

#0=Rural, 1=Urban, more individuals completed primary school in urban areas compared to rural.


```

```python
#What about secondary? (instlevel7 = 1 means completed secondary)

#1st Step: group and filter by HoH
Secondary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel7']].groupby(by=['parentesco1', 'area1', 'instlevel7']).sum().reset_index()

tempdf = (Secondary[Secondary['parentesco1']== 1])
tempdf
#Now filter to only those who completed Secondary school

completed_only = (tempdf[tempdf['instlevel7']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel7', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

#0=Rural, 1=Urban, more individuals completed secondary school in urban areas compared to rural.Also, the disparity between the two is even bigger at this level.
```

```python
#Tertiary?
#Now filter to only those who completed Tertiary school

#1st Step: group and filter by HoH
Tertiary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel8']].groupby(by=['parentesco1', 'area1', 'instlevel8']).sum().reset_index()

tempdf = (Tertiary[Tertiary['parentesco1']== 1])
tempdf
#Now filter to only those who completed Secondary school

completed_only = (tempdf[tempdf['instlevel8']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel8', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

#0=Rural, 1=Urban, more individuals completed secondary school in urban areas compared to rural.Also, the disparity between the two is the greatest at this level.
```
