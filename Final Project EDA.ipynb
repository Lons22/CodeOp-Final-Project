# Costa Rica Proxy Means Test for Poverty Levels: Data Exploration

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 200)
```

```python
data = pd.read_csv('train.csv.zip')
```

```python
data.head()
```

```python
f"The dataset has {data.shape[0]} rows and {data.shape[1]} columns."
```

```python
#Creating a new dataframe

new_data = data[['male', 'female', 'instlevel1','instlevel2','instlevel3', 'instlevel4', 'instlevel5', 'instlevel6', 'instlevel7', 'instlevel8', 'instlevel9', 'rooms','bedrooms', 'hhsize', 'v14a', 'refrig', 'escolari', 'area1', 'area2', 'edjefe', 'edjefa','estadocivil3','estadocivil4','estadocivil5','estadocivil6','estadocivil7', 'parentesco1', 'hogar_total','lugar1',
'lugar2','lugar3','lugar4','lugar5', 'lugar6','v2a1','overcrowding', 'tipovivi1','tipovivi2','tipovivi3','tipovivi4','tipovivi5', 'meaneduc']]
```

```python
new_data.head()
```

# Head of Houshold (HoH) Ratios

```python
#Finding out how many men & women are HoH

household_head = new_data[['parentesco1', 'male', 'female']].groupby(by=['parentesco1']).sum().reset_index()

tempdf = household_head[household_head['parentesco1']== 1] #access column and filtering on its value
 
newdf = pd.melt(tempdf, id_vars=['parentesco1'])

newdf['value'].plot.pie(autopct='%1.1f%%', title = 'Ratio of Male to Female HoH', labels = ['Male', 'Female']);

#based on this result more men are HoH than women.
```

```python
#Rural or Urban?
#area1=1 zona urbana, area2=2 zona rural
new_data['area1'].value_counts(normalize = True).plot.pie(autopct='%1.1f%%', title = 'Ratio of Urban to Rural', labels = ['Urban', 'Rural']);
#Based on this result, most of the data collected is of people living in urban areas.
```

```python
household_head = new_data[['parentesco1', 'male', 'female', 'area1']].groupby(by=['parentesco1', 'area1']).sum().reset_index()
household_head

HoH_filter = (household_head['parentesco1'] == 1) #& (household_head['area1']==1)

tempdf = household_head[HoH_filter] #accss column and filtering on its value
tempdf

newdf = pd.melt(tempdf, id_vars=['parentesco1' , 'area1'])
newdf = newdf.drop('parentesco1', axis = 1)

newdf

ax = newdf['value'].plot.pie(autopct='%1.1f%%', subplots = True, title='HoH Ratio by Area and Gender', labels = ['Rural Males', 'Urban Males', 'Rural Females', 'Urban Females']);

# 0= Rural Males, 1 = Urban Males, 2 = Rural Females, 3 = Urban Females
```

```python
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

plt.title('Number of HoH by Area & Gender')
plt.xlabel('Area')
sns.despine()
plt.show()

#Blue is urban areas, so there are more men HOH compared to female. Orange is rural & there are also more men HoH than feamle.
```

# Education Breakdown

## Primary School

```python
#Overall Primary
overall_primary = new_data[['parentesco1', 'male', 'female', 'instlevel3']].groupby(by=['parentesco1', 'instlevel3']).sum().reset_index()

tempdf = overall_primary[overall_primary['parentesco1']== 1] #access column and filtering on its value
 
newdf = pd.melt(tempdf, id_vars=['parentesco1', 'instlevel3'])
newdf=newdf.drop('parentesco1', axis = 1)

newdf

sns.barplot(x="variable", y="value", hue="instlevel3", data=newdf);

plt.title('Primary School by Gender')
plt.xlabel('Gender')
plt.ylabel('Number')
sns.despine()
plt.show()

# 1=completed primary school, this graph shows that overall people have not completed primary school, of those that did, more males did compared to females.

```

```python
#What Percentage of HoH Completed Primary school by location? #instlevel3 = 1 means completed primary school

#1st Step: group and filter by HoH
Primary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel3']].groupby(by=['parentesco1', 'area1', 'instlevel3']).sum().reset_index()

tempdf = (Primary[Primary['parentesco1']== 1])
tempdf
#Now filter to only those who completed primary school

completed_only = (tempdf[tempdf['instlevel3']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel3', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);


plt.title('Primary School Completion by Area & Gender')
plt.xlabel('Gender')
plt.ylabel('Number')
sns.despine()
plt.show()

#0=Rural, 1=Urban, more individuals completed primary school in urban areas compared to rural.


```

```python
ax = newdf['value'].plot.pie(autopct='%1.1f%%', subplots = True ,title='Ratio of Primary School Completion by Area and Gender', labels = ['Rural Males', 'Urban Males', 'Rural Females', 'Urban Females']);

#
```

## Secondary School

```python
#What about secondary? (instlevel5 = 1 means completed secondary)

#1st Step: group and filter by HoH
Secondary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel5']].groupby(by=['parentesco1', 'area1', 'instlevel5']).sum().reset_index()

tempdf = (Secondary[Secondary['parentesco1']== 1])
tempdf
#Now filter to only those who completed Secondary school

completed_only = (tempdf[tempdf['instlevel5']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel5', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

plt.title('Secondary School Completion by Area & Gender')
plt.xlabel('Gender')
plt.ylabel('Number')
sns.despine()
plt.show()

#0=Rural, 1=Urban, more individuals completed secondary school in urban areas compared to rural.
#Also, the disparity between the two is even bigger at this level.
```

```python
newdf['value'].plot.pie(autopct='%1.1f%%', title='Ratio of Secondary School Completion by Area and Gender', labels = ['Rural Males', 'Urban Males', 'Rural Females', 'Urban Females']);
```

## Tertiary School

```python
#Tertiary?
#Now filter to only those who completed Tertiary school

#1st Step: group and filter by HoH
Tertiary = new_data[['parentesco1', 'male', 'female', 'area1', 'instlevel8']].groupby(by=['parentesco1', 'area1', 'instlevel8']).sum().reset_index()

tempdf = (Tertiary[Tertiary['parentesco1']== 1])
tempdf
#Now filter to only those who completed Secondary school

completed_only = (tempdf[tempdf['instlevel8']== 1])
completed_only 

#Drop both HoH columns & school level columns

tempdf=completed_only.drop('parentesco1', axis = 1)
tempdf

newdf = tempdf.drop('instlevel8', axis =1)
newdf

#Create graphs
newdf = pd.melt(newdf, id_vars=['area1']);
newdf
sns.barplot(x="variable", y="value", hue="area1", data=newdf);

plt.title('Tertiary School Completion by Area & Gender')
plt.xlabel('Gender')
plt.ylabel('Number')
sns.despine()
plt.show()
#0=Rural, 1=Urban, more individuals completed secondary school in urban areas compared to rural.
#Also, the disparity between the two is the greatest at this level.
```

```python
newdf['value'].plot.pie(autopct='%1.1f%%', title='Ratio of Tertiary School Completion by Area and Gender', labels = ['Rural Males', 'Urban Males', 'Rural Females', 'Urban Females']);

#This is to illustrate the disparity of teritary education in rural areas. 
```

# Household Attributes

```python
#Household Attributes

#Do more educated people have bigger houses? (edjefe, edjefa, meaneduc,rooms)

ed = new_data[['edjefe', 'edjefa','meaneduc', 'bedrooms', 'rooms', 'refrig', 'v14a']]
ed

#make a graph
ax = sns.boxplot(data=ed, x="rooms", y="meaneduc");

ax.set(xlabel='Number of Rooms in HH',
       ylabel='Mean Education in years',
       title='Mean years of education vs Number of Rooms HH');
```

```python
ed['rooms'].corr(data['meaneduc'])

#there is a slight correlation to how many years of education a person has and how many bedrooms their house has.
```

```python
#more educated people have a refrigerator in their home

ax= sns.boxplot(data=ed, x="refrig", y="meaneduc")

ax.set(xlabel='Refrigerator in HH',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Refrigerator in HH');
```

```python

ax= sns.boxplot(data=ed, x="v14a", y="meaneduc");

ax.set(xlabel='Toilet in HH',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Toilet in HH');

#more educated people have toilets in their home.
```

```python
#Do more educated people own homes

tipovivi = new_data[['tipovivi1','tipovivi2','tipovivi3','tipovivi4','tipovivi5', 'meaneduc', 'v2a1']]
tipovivi.head()
```

```python

ax= sns.boxplot(data=tipovivi, x="tipovivi1", y="meaneduc");

ax.set(xlabel='Own:Fully Paid off',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Fully Paid Off House');

#Doesn't seem to be a correlation here
```

```python
tipovivi['tipovivi1'].corr(data['meaneduc'])

#no correleation
```

```python
ax= sns.boxplot(data=tipovivi, x="tipovivi2", y="meaneduc");

ax.set(xlabel='Own:Installments',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Paying Installments on House');

#More educated people are paying installments on their house
```

```python
#What about renting?

ax= sns.boxplot(data=tipovivi, x="tipovivi3", y="meaneduc");

ax.set(xlabel='Rent',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Renting');


#slight correlation of more eduated people renting their homes
```

```python
ax= sns.scatterplot(data=tipovivi, x="meaneduc", y="v2a1");

ax.set(xlabel='Mean Education in years',
       ylabel='Monthly rent amount',
       title='Mean Years of Education vs Monthly Rent Amount');

ax.set_ylim(0, 0.50e6)

#higher montlhy rents are correlated with more years of education on average

```

```python
ax= sns.boxplot(data=tipovivi, x="tipovivi4", y="meaneduc");

ax.set(xlabel='Precarious Housing',
       ylabel='Mean Education in years',
       title='Mean Years of Education vs Precarious Housing');

#less educated people are more susceptible to precarious housing (homelessness)
```

```python
#Do less educated people experience more overcrowding?
```

```python
overcrowding = new_data[['overcrowding', 'meaneduc']]

ax= sns.scatterplot(data=overcrowding, x="overcrowding", y="meaneduc");

#there's an inverse relationship here. The people who expereience more people per bedroom have less years of education.
```

# Take Aways

   ### 1. There is a large education disparity within rural and urban areas in Costa Rica. 
   
   ### 2. There are gender disparities also.
   
   ### 3. There is no correlation between the average years of eductation and owning a home. Surprisingly, there is a slight positive correlation between renting and more education on average.
   
   ### 4. Regarding housing attributes, lower education correlates with having a toilet & refrigerator in the home. Also, lower education correlations with an increased likelihood of precarious housing & overcrowding.
